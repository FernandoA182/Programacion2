```python
url = 'https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/'
# Para acceder al URL
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
data=[]
data_iterator = iter(soup.find_all('td'))

while True:
    try:
        pais = next(data_iterator).text
        casos_confirmados = next(data_iterator).text
        muertes = next(data_iterator).text
        region = next(data_iterator).text

        data.append((
            pais,
            int(casos_confirmados.replace(',', '')),
            int(muertes.replace(',', '')),
            region
        ))
    except StopIteration:
        break
data
```

```python
data2=data
def mi_funcion(x):
    return x[2]

data2.sort(key=mi_funcion, reverse=True)

data2
```

```python
#Ordenar por numero de casos confirmados
data.sort(key = lambda x: x[1], reverse=True)

#lambda x: x[1] sustituye a "mi_funcion"

data
```

Ordenando ascendentemente la lista de tuplas con ciclos for

```python
campo_a_ordenar = 0

cant_tuplas = len(data)
print(cant_tuplas)

for i in range(0, cant_tuplas):
    for j in range(0, cant_tuplas-i-1):
        if (data[j][campo_a_ordenar] > data[j + 1][campo_a_ordenar]):
            temp = data[j]
            data[j]= data[j + 1]
            data[j + 1]= temp
        
data
```

### Ejemplos de Ciclos Anidados

Encontrando números repetidos en una lista

#### Ejemplo 1 de Ciclos Anidados

Utilizando cuclos for anidados

```python
num=[4,7,8,9,2,4,1,9]

num_rep=[]

for i in range(len(num)):
    for j in range(len(num)):
        if i != j:
            if num[i] == num[j] and num[i] not in num_rep:
                num_rep.append(num[i])
                
num_rep
```

### Ejemplo de Ciclos Anidados

```python
blusas = ["Blusa Roja", "Blusa Amarilla"]

pantalones = ["Pantalon Negro", "Pantalon Blanco", "Pantalon Azul"]

combinaciones=[]

for i in range(len(blusas)):
    for j in range(len(pantalones)):
        combinaciones.append((blusas[i],pantalones[j]))
        
combinaciones
```

### Listas de Comprensión con Ciclos Anidados

Permiten crear listas en una sola linea  
[(Lista a crear) for iterable]  
[(Lista a crear) for iterable if condicion]

#### Ejemplo 1 de Listas de Compresión con Ciclos Anidados

Sin Lista de Comprensión

```python
lista_pares = []

for i in range(1, 3):
    for j in range(1, 5):
        lista_pares.append((i, j))
        
lista_pares
```

Con lista de Comprensión

```python
lista_pares2 = [(i, j) for i in range(1,3) for j in range(1, 5)]
lista_pares2
```

### Ejercicio 3

Obtén las combinaciones de blusas y pantalones utilizando listas de comprensión

```python
blusas = ["Blusa Roja", "Blusa Amarilla"]

pantalones = ["Pantalon Negro", "Pantalon Blanco", "Pantalon Azul"]

lista_pares3 = [(blusas[i], pantalones[j]) for i in range(len(blusas)) for j in range(len(pantalones))]
lista_pares3
```
