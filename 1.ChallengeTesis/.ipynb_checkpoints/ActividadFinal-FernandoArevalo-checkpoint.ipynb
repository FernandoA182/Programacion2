### 1. General Transit Feed Specification (GTFS)  
El GTFS define un formato común para los horarios del transporte público y la información geográfica asociada para los desarrolladores de aplicaciones. Disponible en especificaciones estáticas y en tiempo real.

Y por su parte General Transit Feed Specification (GTFS) está dividida en las siguientes dos sub-modalidades:  

### 1.1. GTFS Realtime  
GTFS Realtime es una especificación de feed que permite a las agencias de transporte público proporcionar actualizaciones en tiempo real sobre su flota a los desarrolladores de aplicaciones. Es una extensión de GTFS (Especificación de alimentación de tránsito general), un formato de datos abiertos para horarios de transporte público e información geográfica asociada. GTFS Realtime se diseñó teniendo en cuenta la facilidad de implementación, la buena interoperabilidad de GTFS y un enfoque en la información de los pasajeros.

### Plantear una situación que requiera del cálculo de alguna inferencia estadística.  
Dado el proyecto de expansión de Translink, se está analizando el abrir una nueva ubicación para una nueva ruta.  

#### 1. Definir hipotesis
Hipotesis nula: La ruta tendrá alguna ubicación ya en existencia para otra ruta. direction_name 'count' =! 1  
Hipotesis alternativa: La ubicación será nueva. direction_name 'count' sigue una distribución binomial con p = 1  

#### 2. Nivel de significancia (alpha):
alpha = 0.05

#### 3. Prueba estadistica:
Para la distribución binomial, se puede utilizar la prueba binomial o la aproximación normal. Dado que se tiene un tamaño de muestra relativamente grande, la aproximación normal podría ser apropiada.

#### 4. Calcular la prueba estadistica:
Calcular la prueba estadistica basada en el metodo escogido.

#### 5. P-value:
Calcula el p-valor asociado con la prueba estadistica.

#### 6. Desición:
Si el valor p es menor que alfa, rechace la hipótesis nula a favor de la hipótesis alternativa. Si el valor p es mayor o igual que alfa, no rechace la hipótesis nula.

```python
from scipy.stats import binom_test

# Assuming your count data is in a variable named 'data_direction_name'
new_direction_name = 'Burrard Station'
new_direction_name_count = data_direction_name.loc[new_direction_name, 'count']

# Number of trials (sample size)
n_trials = 1  # Since you have one observation

# Number of successes (count != 1)
n_success = int(new_direction_name_count != 1)

# Hypothesis test
p_value = binom_test(n_success, n_trials, p=1)

# Compare p-value to significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Do not reject the null hypothesis.")
    print("p_value = ", p_value)

```

p_value = 1.0 no es menor a alpha = 0.05
### Conclusión:
Basado en los datos, no hay evidencia para concluir que la nueva ubicación será unica.  
El valor p probablemente fue mayor que el nivel de significancia, lo que indica que el recuento observado no es significativamente diferente de lo esperado bajo el supuesto de que la nueva ubicación no es única.

```python
# Create empty lists to store values
directions = []
observed_values = []
expected_values = []
probabilities = []

# Iterate through each direction
for direction in data_direction_name.index:
    # Observed value for the current direction
    observed_value = data_direction_name.loc[direction, 'count']

    # Expected value under the null hypothesis
    expected_value = n_trials * p_null

    # Probability of observing the observed value under the null hypothesis
    p_observed = binom_test(observed_value, n_trials, p=p_null)

    # Append values to the lists
    directions.append(direction)
    observed_values.append(observed_value)
    expected_values.append(expected_value)
    probabilities.append(p_observed)

# Create a table
comparison_table_all = pd.DataFrame({
    'Direction': directions,
    'Observed Value': observed_values,
    'Expected Value': expected_values,
    'Probability': probabilities
})

print(comparison_table_all)
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.DataFrame({
    'Observed': observed_values,
    'Expected': expected_values
}, index=directions)

# Perform chi-square test
chi2, p_chi2, _, _ = chi2_contingency(contingency_table)

# Print the chi-square statistic and p-value
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_chi2}")

# Check for significance
alpha_chi2 = 0.05
if p_chi2 < alpha_chi2:
    print("Reject the null hypothesis (significant).")
else:
    print("Do not reject the null hypothesis (not significant).")
```

### Distribución Binomial:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Assuming counts is your observed counts
observed_counts = data_direction_name['count']

# Fit the Binomial distribution to your data
n_binomial = len(observed_counts)
p_binomial = observed_counts.mean() / n_binomial  # Estimated parameter for Binomial
expected_counts = binom.pmf(np.arange(n_binomial), n_binomial, p_binomial) * len(observed_counts)

# Plot the observed and expected counts
plt.bar(np.arange(len(observed_counts)), observed_counts, label='Observed', alpha=0.7)
plt.plot(np.arange(len(observed_counts)), expected_counts, label='Expected', color='red', linestyle='dashed', marker='o')
plt.xlabel('Count Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Create a DataFrame for observed, expected, and probability
binomial_table = pd.DataFrame({
    'Count': np.arange(n_binomial),
    'Observed': observed_counts,
    'Expected': expected_counts,
    'Probability': binom.pmf(np.arange(n_binomial), n_binomial, p_binomial)
})

print(binomial_table)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom
from scipy.optimize import curve_fit

# Assuming counts is your observed counts
observed_counts = data_direction_name['count']

# Define the negative binomial probability mass function
def negative_binomial_pmf(k, r, p):
    return nbinom.pmf(k, r, p)

# Use curve_fit to estimate parameters
params, covariance = curve_fit(negative_binomial_pmf, np.arange(len(observed_counts)), observed_counts)

# Extract estimated parameters
r, p_negative_binomial = params

# Generate expected counts based on the estimated parameters
expected_counts = nbinom.pmf(np.arange(len(observed_counts)), r, p_negative_binomial) * len(observed_counts)

# Plot the observed and expected counts
plt.bar(np.arange(len(observed_counts)), observed_counts, label='Observed', alpha=0.7)
plt.plot(np.arange(len(observed_counts)), expected_counts, label='Expected', color='red', linestyle='dashed', marker='o')
plt.xlabel('Count Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Generate the probability mass function (PMF) for the negative binomial distribution
probabilities = nbinom.pmf(np.arange(len(observed_counts)), r, p_negative_binomial)

# Create a DataFrame for observed, expected, and probability
negative_binomial_table = pd.DataFrame({
    'Count': np.arange(len(observed_counts)),
    'Observed': observed_counts,
    'Expected': expected_counts,
    'Probability': probabilities
})

# Print the table
print(negative_binomial_table)
```
